# Setup

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow_probability as tfp
import matplotlib.pyplot as plt
import random as rd
import os
from tqdm import tqdm
import tensorflow_addons as tfa
import pandas as pd

plt.style.use(['science','ieee'])
```

# Oracle 

```python

def Oracle(x,y):
    if (x-0.2)**2 + (y-0.6)**2 <= 0.25:
        return 1
    else:
        return 0
'''

def Oracle(x,y):
    if x**3-1.5*x**2+0.75*x-y <= -0.075:
        return 1
    else:
        return 0
'''
```

# Initial training 

```python
taille_train = 10
taille_batch_train = 100
dataset_train = []

for i in range(taille_train):
    batch = [0 for j in range(taille_batch_train)]
    for k in range(len(batch)):
        batch[k] = [rd.uniform(0, 1) for i in range(2)]
    
    batch = tf.convert_to_tensor(batch)
    dataset_train.append(batch)
        
print(len(dataset_train))
print(len(dataset_train[0]))
#print(dataset_train[0])
```

```python
verite = [[] for i in range(taille_train)]
xy_good = [[],[]]
xy_bad = [[],[]]

for step, i in enumerate(dataset_train):
    i = i.numpy()
    for j in i:
        pred = Oracle(j[0],j[1])
        verite[step].append(pred)
    
        if pred == 0:
            xy_bad[0].append(j[0])
            xy_bad[1].append(j[1])

        if pred == 1:
            xy_good[0].append(j[0])
            xy_good[1].append(j[1])
            
print(len(verite))
print(len(verite[0]))
print(len(xy_good[0]))
print(len(xy_bad[0]))
```

```python
plt.plot(xy_good[0],xy_good[1],'og')
plt.plot(xy_bad[0],xy_bad[1],'or')
```

```python
fnc_activation = 'relu'
fnc_activation_output = 'sigmoid'
inputs = tf.keras.Input(shape=(2,), dtype='float32')
x = keras.layers.Dense(100, trainable=True, activation=fnc_activation)(inputs)
x = keras.layers.Dense(100, trainable=True, activation=fnc_activation)(x)
x = keras.layers.Dense(100, trainable=True, activation=fnc_activation)(x)
x = keras.layers.Dense(100, trainable=True, activation=fnc_activation)(x)
x = keras.layers.Dense(100, trainable=True, activation=fnc_activation)(x)
outputs = keras.layers.Dense(1, activation=fnc_activation_output)(x)
model = tf.keras.Model(inputs,outputs)

lr = 1e-3
momentum = 0.05
wd = 0.001

#optimizer = tfa.optimizers.SGDW(learning_rate=lr, momentum = momentum,  weight_decay = wd)
#optimizer = tf.keras.optimizers.SGD(learning_rate=lr, momentum = momentum)
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)

model.summary()



loss_fn = tf.keras.losses.BinaryCrossentropy(from_logits=False, reduction='sum_over_batch_size')
#False pour une probabilité ie [0;1]
```

```python
for step , i in enumerate(dataset_train):
    with tf.GradientTape() as tape:
        logits = model(tf.convert_to_tensor(i))
        
        true = tf.convert_to_tensor([verite[step]])
        true = tf.transpose(true)
        
        loss = loss_fn(true,logits)
    
    gradients = tape.gradient(loss, model.trainable_weights)
    optimizer.apply_gradients(zip(gradients, model.trainable_weights))
```

# Scan 

```python
L_size = 5000
K_size = 100
train_step = 200

a = 0.0001
alpha = 0.005

K_pool = []
K_pred = []

compteur = 0

q = 1

epsilon = 0.3

for step , i in enumerate(range(train_step)):
    print('Step {} / {}'.format(i+1,train_step))
    
    p = 2*min(q,0.5)
    p = round(p,1)
    
    if p == 1:
        K = [[rd.uniform(0,1) for j in range(2)] for k in range(K_size)]
    
    if p < 1:
        pK_size = int(K_size*p)
        pm1 = round((1-p),1)
        
        pK = [[rd.uniform(0,1) for j in range(2)] for k in range(pK_size)]
        
        L = [[rd.uniform(0,1) for j in range(2)] for k in range(L_size)]
        L = tf.convert_to_tensor(L)
        L_logits = model(tf.convert_to_tensor(L))

        s = tf.math.multiply(L_logits,tf.math.subtract(1,L_logits))
        index = tf.math.argmax(s, axis=0).numpy()[0]

        K = []
        K.append(L[index])
        s = tf.concat([s[0:index], s[index+1:len(s)]],0)
        L = tf.concat([L[0:index], L[index+1:len(L)]],0)


        rj_tensor = tf.zeros([len(s),1])
        for j in range(int(pm1*K_size)-1):

            ri = tf.convert_to_tensor(K[-1])
            ri = tf.math.multiply(tf.ones([len(s),1]),ri)

            dabs = tf.math.abs(tf.math.subtract(ri,L))
            dsquared = tf.math.multiply(dabs, dabs)
            dsquared = tf.reduce_mean(dsquared, axis=1)

            rj_list = dsquared.numpy()
            rj_list = np.where(rj_list<0.01**2, -a/(a+rj_list), rj_list*0)
            rj_min = np.amin(rj_list)

            rj_list = tf.convert_to_tensor([rj_list])
            rj_list = tf.transpose(rj_list)
            rj_tensor = tf.math.add(rj_tensor,rj_list)

            sigma = s.numpy().std() 

            S_part1 = tf.math.multiply((1-alpha),s)
            S_part2 = tf.math.multiply((alpha*sigma/(4*rj_min)),rj_tensor)
            S = tf.math.add(S_part1 , S_part2)

            index2 = tf.math.argmax(S, axis=0).numpy()[0]

            K.append(L[index2])

            s = tf.concat([s[0:index2], s[index2+1:len(s)]],0)
            L = tf.concat([L[0:index2], L[index2+1:len(L)]],0)
            rj_tensor = tf.concat([rj_tensor[0:index2], rj_tensor[index2+1:len(rj_tensor)]],0)
            
        K = K + pK
    
    assert len(K) == 100
    K_pool.append(K)
    
    pred_liste = []
    for x,y in K:
        pred = Oracle(x,y)
        pred_liste.append(pred)
    
    K_pred.append(pred_liste)     
    
    batch_pred = [K_pred[-1]]
    all_batch = [K_pool[-1]]
    if compteur == 10:
        compteur = 0
        all_batch = K_pool
        batch_pred = K_pred
    
    for j in range(len(all_batch)): 
        with tf.GradientTape() as tape:
            logits = model(tf.convert_to_tensor(all_batch[j]))

            true = tf.convert_to_tensor([batch_pred[j]])
            true = tf.transpose(true)
            
            true_prime = tf.cast(true, dtype=tf.float32)
            q = tf.math.abs(tf.math.subtract(true_prime,logits))
            q = tf.reduce_mean(q, axis=0)
            q = round(q.numpy()[0],1)
            
            '''
            if step > 1:
                if q > q_old:
                    lr -= lr*epsilon
                    optimizer.learning_rate.assign(lr)
                print(q_old, q, optimizer.learning_rate)
            q_old = q
            '''
            
            loss = loss_fn(true,logits)

        gradients = tape.gradient(loss, model.trainable_weights)
        optimizer.apply_gradients(zip(gradients, model.trainable_weights))
    
    compteur += 1
```

```python
dataset = []
for i in range(len(K_pool)):
    dataset += K_pool[i]

dataset = tf.convert_to_tensor(dataset)
prediction = model(dataset)
```

```python
xr = []
yr = []
pred_r = []
xj = []
yj = []
pred_j = []
xv = []
yv = []
pred_v = []

for indice in tqdm(range(len(prediction))):
    if prediction[indice]<=0.4:
            xr.append(dataset[indice][0])
            yr.append(dataset[indice][1])
            pred_r.append(prediction[indice])
    if prediction[indice]>=0.6:
            xv.append(dataset[indice][0])
            yv.append(dataset[indice][1])
            pred_v.append(prediction[indice])
    if prediction[indice]<0.6 and prediction[indice]>0.4:
            xj.append(dataset[indice][0])
            yj.append(dataset[indice][1])
            pred_j.append(prediction[indice])
```
