```python
def plot_histo(title, xlabel, ylabel, list_hist,bins, label, color='blue', legend = False, histtype='bar', save=False, file_folder=''):
    plt.figure()
    for indice, hist in enumerate(list_hist):
        plt.hist(hist,bins,label=label[indice],histtype=histtype,color=color[indice],stacked=True,linestyle='solid')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    if legend == True:
        plt.legend()
    if save == True:
        plt.savefig(file_folder)
    
def subplot_histo(titre, xlabel, ylabel, bins, x, y, share, subtitle, hist_list, label, color='blue', legend = False, histtype='bar', save=False, file_folder=''):  
    fig , ax = plt.subplots(x, y, sharex=share[0], sharey=share[1], constrained_layout = True)
    plt.suptitle(titre)
    fig.text(0.5, -0.05, xlabel, ha='center')
    fig.text(-0.04, 0.5, ylabel, va='center', rotation='vertical')

    for indice, hist in enumerate(hist_list): 
        compteur = 0
        for i in range(x):
            if y != 1:
                for j in range(y):
                    ax[i, j].hist(hist[compteur],bins,histtype=histtype,label=label[indice], color=color[indice], stacked=True, linestyle='solid')
                    ax[i, j].set_title(subtitle[compteur])
                    compteur+=1
                    if legend == True:
                        ax[i, j].legend()
            else:
                ax[i].hist(hist[compteur],bins,histtype=histtype, label=label[indice], color=color[indice], stacked=True, linestyle='solid')
                ax[i].set_title(subtitle[compteur])
                compteur+=1
                if legend == True:
                    ax[i].legend()
    if save == True:
        plt.savefig(file_folder)
```

```python
#file = 'K_pool_tot.csv'
file = 'K_pool_tot.csv'
with open(path+folder_r+file) as f:
    f.readline()
    data = csv.reader(f, delimiter='\t')
    data = list(data)
    
good_data_logits = [[] for i in range(42)]
bad_data_logits  = [[] for i in range(42)]
```

```python
for i in tqdm(data):
    if i[-1] == '0':
        for j in range(42):
            bad_data_logits[j].append(float(i[j]))
    
    if i[-1] == '1':
        for j in range(42):
            good_data_logits[j].append(float(i[j]))
            
good_data_logits = tf.convert_to_tensor(good_data_logits)
bad_data_logits = tf.convert_to_tensor(bad_data_logits)
```

```python
#folder_model = '2022-06-08-08:57:07/'
folder_model = folder_r
model_name = 'model_II.h5'

model = tf.keras.models.load_model(path+folder_model+model_name)
model.summary()
```

```python
good_data_logits = tf.transpose(good_data_logits)
good_logits_pred = model(good_data_logits)

bad_data_logits = tf.transpose(bad_data_logits)
bad_logits_pred = model(bad_data_logits)
```

```python
good_compteur_1 = 0
good_compteur_0 = 0
good_compteur_05 = 0

for i in good_logits_pred:
    if i.numpy() < 0.5:
        good_compteur_0 += 1
    if i.numpy() > 0.5:
        good_compteur_1 += 1
    if i.numpy() == 0.5:
        good_compteur_05 += 1
```

```python
bad_compteur_1 = 0
bad_compteur_0 = 0
bad_compteur_05 = 0

for i in bad_logits_pred:
    if i.numpy() < 0.5:
        bad_compteur_0 += 1
    if i.numpy() > 0.5:
        bad_compteur_1 += 1
    if i.numpy() == 0.5:
        bad_compteur_05 += 1
```

```python
print(bad_compteur_1,bad_compteur_0,bad_compteur_05)
print(good_compteur_1,good_compteur_0,good_compteur_05)

print(good_compteur_1+good_compteur_0+good_compteur_05)
print(good_compteur_1+good_compteur_0)

print(bad_compteur_1+bad_compteur_0+bad_compteur_05)
print(bad_compteur_1+bad_compteur_0)
```

```python
TP = good_compteur_1
FN = good_compteur_0

FP = bad_compteur_1
TN = bad_compteur_0
```

```python
print('True positive: ',TP)
print('False positive: ',FP)
print('True negative: ',TN)
print('False negative: ',FN)

print('Positive prediction: ',round(TP/(TP+FP)*100,2))
print('Negative prediction: ',round(TN/(TN+FN)*100,2))

print('Sensitivity: ',round(TP/(TP+FN)*100,2))
print('Specificity: ',round(TN/(FP+TN)*100,2))
```
