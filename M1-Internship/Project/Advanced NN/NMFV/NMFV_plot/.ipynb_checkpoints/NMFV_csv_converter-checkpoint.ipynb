```python
def write_slha(NMFV, nom_slha):
    f = open(nom_slha, "w")
    
    f.write("BLOCK MODSEL \n")
    f.write(" 1    0                               #MSSM \n")
    f.write(" 6    1                               #flavour violation \n")
                                                    
    f.write("BLOCK SMINPUTS \n")                    
    f.write("  1   1.27931417e+02                  # alpha^(-1) SM MSbar(M1) \n")
    f.write("  2   1.663900000e-05                 # G_Fermi \n")
    f.write("  3   0.119                           # alphas(MS) SM MSbar \n")
    f.write("  4   91.1876                         # MZ(pole) \n")
    f.write("  5   4.18                            # mb(mb) SM MSbar \n")
    f.write("  6   172.9                           # mtop(pole) \n")
    f.write("  7   1.776860000e+00                 # m mtau(pole) \n")
                                                      
    f.write("BLOCK MINPAR \n")                     
    f.write("  3   "+str(NMFV[0])+"                # tanb \n")
    f.write("  4   1                               # sign mu \n")
                                                   
    f.write("BLOCK EXTPAR \n")                    
    f.write("  0    "+str(-1.0)+"                  # Q \n")
    f.write("  1    "+str(NMFV[1])+"               # M1 \n")
    f.write("  2    "+str(NMFV[2])+"               # M2 \n")
    f.write("  3    "+str(NMFV[3])+"               # M3 \n")
    f.write("  23   "+str(NMFV[4])+"               # mu \n")
    f.write("  26   "+str(NMFV[5])+"               # mA \n")
    f.write("  25   "+str(NMFV[0])+"               # tanb \n")
    
    f.write("BLOCK MSQ2IN \n")                            
    f.write("  1  1    "+str(NMFV[6])+"            # mq2_11 \n")
    f.write("  2  1    "+str(0)+"                  # mq2_21 \n")
    f.write("  3  1    "+str(0)+"                  # mq2_31 \n")
    f.write("  1  2    "+str(0)+"                  # mq2_12 \n")
    f.write("  2  2    "+str(NMFV[7])+"            # mq2_22 \n")
    f.write("  3  2    "+str(0)+"                  # mq2_32 \n")
    f.write("  1  3    "+str(0)+"                  # mq2_13 \n")
    f.write("  2  3    "+str(NMFV[8])+"            # mq2_23 \n")
    f.write("  3  3    "+str(NMFV[9])+"            # mq2_33 \n")
    
    f.write("BLOCK MSU2IN \n")                            
    f.write("  1  1    "+str(NMFV[10])+"           # mu2_11 \n")
    f.write("  2  1    "+str(0)+"                  # mu2_21 \n")
    f.write("  3  1    "+str(0)+"                  # mu2_31 \n")
    f.write("  1  2    "+str(0)+"                  # mu2_12 \n")
    f.write("  2  2    "+str(NMFV[11])+"           # mu2_22 \n")
    f.write("  3  2    "+str(0)+"                  # mu2_32 \n")
    f.write("  1  3    "+str(0)+"                  # mu2_13 \n")
    f.write("  2  3    "+str(0)+"                  # mu2_23 \n")
    f.write("  3  3    "+str(NMFV[12])+"           # mu2_33 \n")
    
    f.write("BLOCK MSD2IN \n")                            
    f.write("  1  1    "+str(NMFV[13])+"           # md2_11 \n")
    f.write("  2  1    "+str(0)+"                  # md2_21 \n")
    f.write("  3  1    "+str(0)+"                  # md2_31 \n")
    f.write("  1  2    "+str(0)+"                  # md2_12 \n")
    f.write("  2  2    "+str(NMFV[14])+"           # md2_22 \n")
    f.write("  3  2    "+str(0)+"                  # md2_32 \n")
    f.write("  1  3    "+str(0)+"                  # md2_13 \n")
    f.write("  2  3    "+str(NMFV[15])+"            # md2_23 \n")
    f.write("  3  3    "+str(NMFV[16])+"           # md2_33 \n")
    
    f.write("BLOCK MSL2IN \n")                            
    f.write("  1  1    "+str(NMFV[17])+"           # ml2_11 \n")
    f.write("  2  1    "+str(0)+"                  # ml2_21 \n")
    f.write("  3  1    "+str(0)+"                  # ml2_31 \n")
    f.write("  1  2    "+str(0)+"                  # ml2_12 \n")
    f.write("  2  2    "+str(NMFV[18])+"           # ml2_22 \n")
    f.write("  3  2    "+str(0)+"                  # ml2_32 \n")
    f.write("  1  3    "+str(0)+"                  # ml2_13 \n")
    f.write("  2  3    "+str(0)+"                  # ml2_23 \n")
    f.write("  3  3    "+str(NMFV[19])+"           # ml2_33 \n")
    
    f.write("BLOCK MSE2IN \n")                            
    f.write("  1  1    "+str(NMFV[20])+"           # me2_11 \n")
    f.write("  2  1    "+str(0)+"                  # me2_21 \n")
    f.write("  3  1    "+str(0)+"                  # me2_31 \n")
    f.write("  1  2    "+str(0)+"                  # me2_12 \n")
    f.write("  2  2    "+str(NMFV[21])+"           # me2_22 \n")
    f.write("  3  2    "+str(0)+"                  # me2_32 \n")
    f.write("  1  3    "+str(0)+"                  # me2_13 \n")
    f.write("  2  3    "+str(0)+"                  # me2_23 \n")
    f.write("  3  3    "+str(NMFV[22])+"           # me2_33 \n")
    
    f.write("BLOCK TEIN \n")                            
    f.write("  1  1    "+str(0)+"                  # T_e_11 \n")
    f.write("  2  1    "+str(0)+"                  # T_e_21 \n")
    f.write("  3  1    "+str(0)+"                  # T_e_31 \n")
    f.write("  1  2    "+str(0)+"                  # T_e_12 \n")
    f.write("  2  2    "+str(0)+"                  # T_e_22 \n")
    f.write("  3  2    "+str(0)+"                  # T_e_32 \n")
    f.write("  1  3    "+str(0)+"                  # T_e_13 \n")
    f.write("  2  3    "+str(0)+"                  # T_e_23 \n")
    f.write("  3  3    "+str(NMFV[41])+"           # T_e_33 \n")
    
    f.write("BLOCK TUIN \n")                            
    f.write("  1  1    "+str(NMFV[23])+"           # T_u_11 \n")
    f.write("  1  2    "+str(NMFV[24])+"           # T_u_12 \n")
    f.write("  1  3    "+str(NMFV[25])+"           # T_u_13 \n")
    f.write("  2  1    "+str(NMFV[26])+"           # T_u_21 \n")
    f.write("  2  2    "+str(NMFV[27])+"           # T_u_22 \n")
    f.write("  2  3    "+str(NMFV[28])+"           # T_u_23 \n")
    f.write("  3  1    "+str(NMFV[29])+"           # T_u_31 \n")
    f.write("  3  2    "+str(NMFV[30])+"           # T_u_32 \n")
    f.write("  3  3    "+str(NMFV[31])+"           # T_u_33 \n")
    
    f.write("BLOCK TDIN \n")                            
    f.write("  1  1    "+str(NMFV[32])+"           # T_d_11 \n")
    f.write("  1  2    "+str(NMFV[33])+"           # T_d_12 \n")
    f.write("  1  3    "+str(NMFV[34])+"           # T_d_13 \n")
    f.write("  2  1    "+str(NMFV[35])+"           # T_d_21 \n")
    f.write("  2  2    "+str(NMFV[36])+"           # T_d_22 \n")
    f.write("  2  3    "+str(NMFV[37])+"           # T_d_23 \n")
    f.write("  3  1    "+str(NMFV[38])+"           # T_d_32 \n")
    f.write("  3  2    "+str(NMFV[39])+"           # T_d_32 \n")
    f.write("  3  3    "+str(NMFV[40])+"           # T_d_33 \n")
    
    f.close()
```
